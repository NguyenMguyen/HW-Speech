```python
from scipy.signal import tf2zpk
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
b = [1, 0, 0]
a = [1., -1., -1.]
(z,p,k) = tf2zpk(b,a)
```

```python
print(z, p, k)
```

```python
circle1 = plt.Circle(z, 1., fill=False, color='r')
circle2 = plt.Circle(z, 0.05, fill=False, color='b')

fig, axes = plt.subplots(figsize=(4,4)) 
plt.plot(p, [0,0], 'gx', color='b')
axes.set_aspect( 1 )
axes.add_artist( circle1 )
axes.add_artist(circle2)
axes.set_xlim((-2, 2))
axes.set_ylim((-2, 2))

plt.xticks(range(-2,3,1))
plt.yticks(range(-2,3,1))
plt.tight_layout()
plt.grid()
plt.savefig('bai1d.png', dpi=500)
plt.show()
```

```python
omega = np.linspace(0, 2*math.pi,256)
a=0.87
H1 = 1/(1-a*np.exp(-1j*omega))
fig, ax = plt.subplots(figsize=(12,4))

plt.subplots_adjust(wspace=.5, hspace=.5)
ax.plot(omega, abs(H1), linewidth=2)
ax.set_xticks([0, np.pi, 2*np.pi])
ax.set_yticks(range(0,9,2))
plt.xlim(0, 2*np.pi)
ax.set_xticklabels(['0', r'$\pi$', r'2$\pi$'])
ax.set_ylabel('Magnitude')
ax.set_xlabel('Frequency')
ax.set_title('Magnitude Response')

plt.grid()
plt.savefig('bai2.png',dpi=300)
plt.show()
```

```python
import pylab
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
import wave
import sys
from scipy import signal
import librosa , librosa.display
from scipy.io.wavfile import read
```

```python
def drawSignal(file):
    spf = wave.open(file, "r")

    # Extract Raw Audio from Wav File
    signal = spf.readframes(-1)
    signal = np.frombuffer(signal, "int16")

    fs = spf.getframerate()
    Time = np.linspace(0, len(signal) / fs * 1000, num=len(signal))
    plt.figure(figsize = (16, 4))
    plt.ylabel('Biên độ')
    plt.xlabel('Time in ms')
    plt.title("Xe.wav, Fs=16000Hz")
    plt.plot(Time, signal, c='black')
    plt.grid(c='blue', linestyle='--')
    plt.yticks([-10000, 0, 10000])
    plt.tight_layout()
    plt.savefig('bai1.png', dpi=600)
    plt.show()

drawSignal("Xe.wav")
```

```python
samplingFrequency, signalData = wf.read("Xe.wav")
fig, ax = plt.subplots(2, 1, figsize=(16, 7))

plt.subplot(211)
plt.specgram(signalData, Fs=(samplingFrequency/1000), cmap='Greys')
plt.xlabel('Time')
plt.ylabel('kHz')

plt.subplot(212)
spf = wave.open('Xe.wav', "r")
signal = spf.readframes(-1)
signal = np.frombuffer(signal, "int16")
fs = spf.getframerate()
Time = np.linspace(0, len(signal) / fs, num=len(signal))
plt.xlim(0., 0.8)
plt.ylim(np.min(signal), np.max(signal))
plt.yticks([np.min(signal), np.max(signal)])
plt.xlabel('Time')
plt.plot(Time, signal, c='black')

plt.tight_layout()
plt.savefig('bai2.png', dpi=600)
plt.show()
```
